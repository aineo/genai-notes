###Load models for inference

```python
import diffusers
import torch

pipe = diffusers.MarigoldDepthPipeline.from_pretrained(
    "prs-eth/marigold-depth-lcm-v1-0", variant="fp16", torch_dtype=torch.float16
).to("cuda")

image = diffusers.utils.load_image("Lenna.png")
depth = pipe(image)
```

###Depth, Depth16bit, Surface normals

```python
vis = pipe.image_processor.visualize_depth(depth.prediction)
vis[0].save("Lenna_depth.png")
vis

depth_16bit = pipe.image_processor.export_depth_to_16bit_png(depth.prediction)
depth_16bit[0].save("Lenna_depth_16bit.png")
depth_16bit
```

###Surface normals

```python
pipe_normals = diffusers.MarigoldNormalsPipeline.from_pretrained(
    "prs-eth/marigold-normals-lcm-v0-1", variant="fp16", torch_dtype=torch.float16
).to("cuda")

image = diffusers.utils.load_image("Lenna.png")
normals = pipe_normals(image)

vis_normals = pipe_normals.image_processor.visualize_normals(normals.prediction)
vis_normals[0].save("Lenna_normals.png")
```
